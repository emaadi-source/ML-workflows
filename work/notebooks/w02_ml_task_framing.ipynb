## 1. My lane as an ML task (type)

*Classification, clustering, ranking, or scoring — which one, and why?*



The **Refresh / Content Opportunity Scoring** lane is primarily a **ranking and scoring** task. The goal is to assign each content page a priority score and rank pages according to their need for review. Instead of simply predicting whether a page is declining, the system helps prioritize which pages should be reviewed first when time and resources are limited.

The output is a ranked list of content pages that supports decision-making for SEO and content teams. Ranking is more appropriate than simple classification because the objective is to order pages by review priority rather than only assigning a yes/no label.

## 2. Target or proxy

*What would you predict? Where does that label come from — observed outcome or a defined rule?*



The target for this project is the **is_declining_label**, which indicates whether a content page is currently classified as declining. This label is derived from the page's trend direction and is therefore a **proxy**, not a directly observed future outcome.

The starter dataset uses this proxy to demonstrate the complete ML workflow. In a more advanced version of the project, the target would ideally be an observed future outcome, such as whether a page declines or recovers during a later time window.

## 3. Success metric

*One metric you can defend. What number means 'good'?*



The primary success metric for this project is **Precision@K**. Since the output is a ranked list of content pages, Precision@K measures how many of the top-ranked pages are actually good candidates for review.

For example, if the SEO team reviews only the top 20 recommended pages, a higher Precision@20 means that more of those pages truly deserved attention. This metric matches the real business decision because content teams usually have limited time and can only review a small number of pages at a time.

## 5. Why ML beats a fixed rule here

*What makes the pattern too messy for an if-statement?*



A fixed rule can only evaluate a small number of conditions, such as recommending a page for review if its impressions are above a certain threshold and its trend is declining. While simple rules are easy to understand, they cannot capture the complex relationships between multiple performance signals.

Machine learning can analyze many features simultaneously, including impressions, clicks, CTR, sessions, average position, content age, engagement rate, and trend information. By learning patterns from these combined signals, ML can rank pages more effectively than a single fixed rule and provide better decision support for content teams.

The purpose of the model is not to replace human judgment but to prioritize pages so that reviewers spend their time on the most promising opportunities.
